```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
a =np.linspace(0,5,10)
b = a**3
```

```python
figure, axes = plt.subplots(1,2)
#axes.plot(a,b)
```

```python
figure, axes = plt.subplots(nrows=2,ncols=4)
```

```python
#оступ к графикам по циклу
figure, axes = plt.subplots(1,2)

for current_ax in axes:
    current_ax.plot(a,b)
```

```python
figure, axes = plt.subplots(1,2)
axes[0].plot(a,b)

axes[0].set_xlabel('a array values')
axes[0].set_ylabel('b array values')
axes[0].set_title('Axes[0] title')

axes[1].plot(b,a)

axes[1].set_xlabel('a array values')
axes[1].set_ylabel('b array values')
axes[1].set_title('Axes[1] title')
```

```python
figure = plt.figure(figsize=(4,3),dpi=100)
axes = figure.add_axes([0,0,1,1])
axes.plot(a,b)
```

```python
figure,axes = plt.subplots(figsize=(4,3),dpi=100)
axes.plot(a,b)
```

```python
figure,axes = plt.subplots(nrows=2,ncols=1,figsize=(4,3),dpi=100)
axes[0].plot(a,b)
axes[1].plot(b,a)
```
