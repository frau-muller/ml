```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
a =np.linspace(0,5,10)
b = a**3
```

```python
figure = plt.figure()
axes = figure.add_axes([0,0,1,1])
axes.plot(a,b,color = 'green',linewidth=5,alpha = 0.3)
#color можно и16-ной системе
#linewidth(lw) -толщина
#alpha прозрачность
```

```python
figure = plt.figure()
axes = figure.add_axes([0,0,1,1])
axes.plot(a,b,color = 'green',linewidth=5,alpha = 0.3,linestyle='--')
```

```python
figure = plt.figure()
axes = figure.add_axes([0,0,1,1])
axes.plot(a,b,color = 'green',linewidth=1,alpha = 0.3,linestyle='--',marker='<',
          markersize=20,markerfacecolor='red',markeredgewidth=3,
          markeredgecolor ='blue')
#толщина кромки markeredgewidth
#получение части графика по оси x и н
axes.set_xlim([0,1])
axes.set_ylim([0,3])
```
