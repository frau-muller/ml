```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
a =np.linspace(0,5,10)
b = a**3
```

```python
plt.plot(a,b)
#дато название осям
plt.xlabel('a array')
plt.ylabel('b array')
#Заголовок
plt.title('Plot for a and b arrays')
```

```python
plt.subplot(1,2,1)
plt.plot(a,b,'r')#r -цвет
plt.subplot(1,2,2)
plt.plot(b,a,'b')
```

```python
figure = plt.figure()
axes = figure.add_axes([0.1,0.1,0.8,0.8])
axes.plot(a,b)
#название осей
axes.set_xlabel('a array values')
axes.set_ylabel('b array values')
axes.set_title('Some title')
```
