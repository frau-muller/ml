```python
import seaborn as sns
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
#pairplot(делает всё автомат.) упрoщенная версия PairGrid,
sns.pairplot(iris)
```

```python
sns.PairGrid(iris)
```

```python
import matplotlib.pyplot as plt
grid = sns.PairGrid(iris)
grid.map(plt.scatter)#передаем тип графика
```

```python
import matplotlib.pyplot as plt
grid = sns.PairGrid(iris)
grid.map_diag(sns.histplot)
grid.map_upper(plt.scatter)
grid.map_lower(sns.kdeplot)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
grid = sns.FacetGrid(data=tips,col='time',row='smoker')
grid.map(sns.histplot,'total_bill')
```

```python
grid = sns.FacetGrid(data=tips,col='time',row='smoker')
grid.map(plt.scatter,'total_bill','tip')
```
