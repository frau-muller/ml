```python
import seaborn as sns
%matplotlib inline
```

```python
dots = sns.load_dataset('dots')

```

```python
dots.head()
```

```python
sns.displot(dots['firing_rate'],kde=True,bins=30)
```

```python
#x i y два столбца,которые мы хотим сравнить между собой из датасет
sns.jointplot(x='coherence',y='firing_rate',data=dots)

```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.displot(tips['total_bill'],kde=True,bins=30)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')

```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.pairplot(tips,hue='sex',palette='husl')

```

```python
sns.rugplot(tips['total_bill'])
```
